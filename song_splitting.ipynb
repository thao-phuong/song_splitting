```python
import youtube_dl
import librosa
import IPython.display as ipd
from IPython.display import Audio
```

```python
url = "https://www.youtube.com/watch?v=unZFadh0Xpg" 
```

```python
ydl_params = {
    'format': 'bestaudio/best',
    'outtmpl': 'chosen_song.wav'
}

with youtube_dl.YoutubeDL(ydl_params) as ytd:
    info = ytd.extract_info(url, download=False)
    status = ytd.download([url])
```

```python
au, rate = librosa.load('chosen_song.wav', sr=11025, mono=False)
```

```python
display(Audio(au[:, 0*rate:40*rate], rate=rate))
```
